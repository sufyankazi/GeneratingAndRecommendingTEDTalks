<h3>Cleaning For LSTM</h3>

```python
LSTMSubs=[]
LSTMString=''
for i in range(len(AllSubs)):
    tempstring=AllSubs[i]
    tempstring=tempstring.replace('\n', ' ')
    tempstring=tempstring.replace('\\', ' ')
    for char in '0123456789:':
        tempstring=re.sub(char,'',tempstring)
    LSTMSubs.append(tempstring)
    tempstring=tempstring.replace('  ', ' ')
    LSTMString=(LSTMString+tempstring).strip()+'\n'
LSTMString[:1000]
```

<h3>Cleaning for other NLP</h3>

```python
ProcessedSubs=[]
SubString=''
LemmaString=''
for i in range(len(AllSubs)):
    tempstring=AllSubs[i]
    tempstring=tempstring.replace('\n', ' ')
    for char in '0123456789!:",;-':
        tempstring=re.sub(char,'',tempstring)
    tempstring=re.sub('\?','',tempstring)
    tempstring=re.sub('\.','',tempstring)
    ProcessedSubs.append(tempstring)
    tempstring=tempstring.replace('\'', '')
    tempstring=tempstring.replace('  ', ' ')
    SubString=(SubString+tempstring).strip()+'\n'
    SubString=SubString.lower()
# l=nltk.wordnet.WordNetLemmatizer()
# for word in SubString:
#     LemmaString=LemmaString+' '+l.lemmatize(word)
SubString[:1000]
SubStopList=SubString.split('\n')
SubList=[]
StopWords=nltk.corpus.stopwords.words("english")
for i in range(len(SubStopList)):
    templist=[]
    for word in SubStopList[i].split():    
        if not word in StopWords:
            templist.append(word)
        SubList.append(templist)
```

<h1>Modeling</h1>

```python
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
WV = gensim.models.Word2Vec(iter=10, workers=8)  # an empty model, no training yet
WV.build_vocab(SubList)  # can be a non-repeatable, 1-pass generator
WV.train(SubList, total_words=WV.corpus_count, epochs=WV.iter)  # can be a non-repeatable, 1-pass generator
```

```python
def tsne_plot(model):
    "Creates and TSNE model and plots it"
    labels = []
    tokens = []

    for word in model.wv.vocab:
        tokens.append(model[word])
        labels.append(word)
    
    tsne_model = TSNE(perplexity=40, n_components=2, init='pca', n_iter=2500, random_state=23)
    new_values = tsne_model.fit_transform(tokens)

    x = []
    y = []
    for value in new_values:
        x.append(value[0])
        y.append(value[1])
        
    plt.figure(figsize=(16, 16)) 
    for i in range(len(x)):
        plt.scatter(x[i],y[i])
        plt.annotate(labels[i],
                     xy=(x[i], y[i]),
                     xytext=(5, 2),
                     textcoords='offset points',
                     ha='right',
                     va='bottom')
    plt.show()
```
